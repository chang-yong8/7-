```python
# 성적 관리 프로그램

# data
# create table > 만들기
# column : 수험번호, 이름, 문항별 정답여부
# insert into values() > 값 넣기

# oracledb
# connect, cursor, execute, commit, close

# flow

# 암호 : 정규표현식 (관리자 : m020202코드, 수험자 : a0202020수험번호 [m])
# 관리자
# 삭제 : 수험번호 > 관리자 프로그램 사용 시 > 김동주
# 수정 : 수험번호, 정답여부 고치기 > 배계현
# 수험자
# 조회 : > (이름,수험번호)출력 고르기 > 수험번호로 찾아서 조회 (성적/전체평균, 합/불) > 김창용
# 오답확인(Y/N) > 번호(num) > 해당 문항의 keyword 같이 출력 > '연습문제풀기(Y/N) > sleep(시간 재기) / 정답, 오답 출력(while/1회)' > 인
# 문항별 정답률 조회 : 문항번호로 정답률 출력 > 인


data = [{'name' : 'A',1 : 'O',2 : 'O',3 : 'O',4 : 'X',5 : 'O'},
        {'name' : 'B',1 : 'O',2 : 'O',3 : 'O',4 : 'X',5 : 'O'},
        {'name' : 'C',1 : 'O',2 : 'O',3 : 'O',4 : 'X',5 : 'O'},
        {'name' : 'D',1 : 'X',2 : 'X',3 : 'O',4 : 'X',5 : 'O'},
        {'name' : 'E',1 : 'O',2 : 'O',3 : 'O',4 : 'X',5 : 'O'}]

keyword = {1 : '고조선', 2 : '삼국', 3 : '고려', 4 : '조선', 5 : '근대'}

while True:
    go = '''
            고르세요.\n
            1. 정답률 확인\n
            2. 내 성적 확인\n
            3. 개인 성적\n
            4. 종료'''
    menu = input(go)
    if menu == '1':
        num = int(input('문제번호를 고르시오'))
        count = 0
        for i in data:
            if i[num] == 'O':
                count += 1
        print(f'정답률 : {count*100/len(data)} %, {keyword[num]}')
    elif menu == '2':
        name = input('이름을 입력하세요')
        for i in data:
            if i['name'] == name:
                print(i)
    elif menu == '3':
        name = input('이름을 입력하세요')
        name_count = 0
        count = 0
        for i in range(len(data)):
            for j in data[i].values():
                if j == 'O':
                    count += 1
        for i in data:
            if i['name'] == name:
                for j in i.values():
                    if j == 'O':
                        name_count += 1
        print(f'{name} : {name_count} 평균 {round(count/len(data))} 개')
    elif menu == '4':
        print('종료합니다.')
        break
    else:
        print('다시 입력하세요')
```
